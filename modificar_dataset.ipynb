Imports

```python
import pandas as pd
import numpy as np
```

Leer archivo

```python
df = pd.read_csv("data/covid_data.csv")
df.head()
```

ModificaciÃ³n de variables

```python
# En vez de fallecido o no, cambio la variable a 'zombificado'. Esta serÃ¡ nuestra variable objetivo
df['sobreviviente'] = df['DATE_DIED'].apply(lambda x: 0 if x != '9999-99-99' else 1)
df.drop(columns=['PREGNANT'], inplace=True)
```

Renombro columnas

```python
df = df.rename(columns={
    'AGE': 'edad',
    'SEX': 'genero',
    'INTUBED': 'mordido',
    'PNEUMONIA': 'infeccion_zombie',
    'INMSUPR': 'inmunidad_baja',
    'ICU': 'unidad_critica',
    'OBESITY': 'movilidad_reducida',
    'TOBACCO': 'resistencia_baja',
    'DATE_DIED': 'fecha_zombificacion',
    'PATIENT_TYPE': 'tipo_contacto',
    'USMER': 'zona_infectada',
    'MEDICAL_UNIT': 'unidad_refugio',
    'CLASIFFICATION_FINAL': 'grupo_supervivencia',
    'DIABETES': 'condicion_base_diabetes',
    'COPD': 'condicion_base_epoc',
    'ASTHMA': 'condicion_base_asma',
    'HIPERTENSION': 'condicion_base_hipertension',
    'OTHER_DISEASE': 'otras_condiciones',
    'CARDIOVASCULAR': 'condicion_base_cardio',
    'RENAL_CHRONIC': 'condicion_base_renal'
})
```

Me creo una copia del dataset original para trabajarlo y modificarlo

```python
df_zombie = df.copy()
df_zombie.head()
```

Modifico columnas redundantes y creo nuevas variables

```python
condiciones_estres = [
    'condicion_base_diabetes', 'condicion_base_epoc',
    'condicion_base_asma', 'condicion_base_hipertension',
    'otras_condiciones', 'condicion_base_cardio',
    'condicion_base_renal'
]

df_zombie[condiciones_estres] = df_zombie[condiciones_estres].replace(2, 0)
df_zombie['estado_mental'] = df_zombie[condiciones_estres].sum(axis=1).apply(lambda x: 0 if x >= 3 else 1) # 0 = alterado, 1 = estable

df_zombie.drop(columns=condiciones_estres, inplace=True) # Borro las columnas para simplificar el dataset

tipos_armas = [0, 1, 2, 3, 4, 5] # Luego serÃ¡n cambiados por texto

np.random.seed(42) # Simular que el 40% tiene arma de defensa
df_zombie['arma_defensa'] = np.random.choice(
    tipos_armas,
    size=len(df_zombie),
    p=[0.4, 0.2, 0.15, 0.1, 0.1, 0.05]
)
```

Cambiar valores por 0 (como deberÃ­an haber estado en un principio)

```python
df_zombie['zona_infectada'] = df_zombie['zona_infectada'].replace(2, 0)
df_zombie['tipo_contacto'] = df_zombie['tipo_contacto'].replace(2, 0)
df_zombie['mordido'] = df_zombie['mordido'].replace(2, 0)
df_zombie['mordido'] = df_zombie['mordido'].replace(97, -1)
df_zombie['mordido'] = df_zombie['mordido'].replace(99, -1)
df_zombie['infeccion_zombie'] = df_zombie['infeccion_zombie'].replace(2, 0)
df_zombie['infeccion_zombie'] = df_zombie['infeccion_zombie'].replace(99, -1)
df_zombie['inmunidad_baja'] = df_zombie['inmunidad_baja'].replace(2, 0)
df_zombie['inmunidad_baja'] = df_zombie['inmunidad_baja'].replace(98, -1)
df_zombie['movilidad_reducida'] = df_zombie['movilidad_reducida'].replace(2, 0)
df_zombie['movilidad_reducida'] = df_zombie['movilidad_reducida'].replace(98, -1)
df_zombie['resistencia_baja'] = df_zombie['resistencia_baja'].replace(2, 0)
df_zombie['resistencia_baja'] = df_zombie['resistencia_baja'].replace(98, -1)
df_zombie['unidad_critica'] = df_zombie['unidad_critica'].replace(2, 0)
df_zombie['unidad_critica'] = df_zombie['unidad_critica'].replace(97, -1)
df_zombie['unidad_critica'] = df_zombie['unidad_critica'].replace(99, -1)
df_zombie['sobreviviente'] = df_zombie['sobreviviente'].replace(2, 0)
# Unifico los refugios
df_zombie['unidad_refugio'] = df_zombie['unidad_refugio'].replace(4, 1)
df_zombie['unidad_refugio'] = df_zombie['unidad_refugio'].replace(10, 1)
df_zombie['unidad_refugio'] = df_zombie['unidad_refugio'].replace(11, 1)
df_zombie['unidad_refugio'] = df_zombie['unidad_refugio'].replace(7, 2)
df_zombie['unidad_refugio'] = df_zombie['unidad_refugio'].replace(8, 2)
df_zombie['unidad_refugio'] = df_zombie['unidad_refugio'].replace(13, 2)
df_zombie['unidad_refugio'] = df_zombie['unidad_refugio'].replace(5, 3)
df_zombie['unidad_refugio'] = df_zombie['unidad_refugio'].replace(6, 3)
df_zombie['unidad_refugio'] = df_zombie['unidad_refugio'].replace(9, 4)
df_zombie['unidad_refugio'] = df_zombie['unidad_refugio'].replace(12, 4)

df_zombie['fecha_zombificacion'] = df_zombie['fecha_zombificacion'].replace('9999-99-99', '0')
```

Guardo archivo final

```python
df_zombie.to_csv("data/zombie_data_binary.csv", index=False)
```
